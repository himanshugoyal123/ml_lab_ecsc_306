```python
import tensorflow as tf
tf.reset_default_graph()
a=[]
b=[]
n=0
t1=0
t2=0
m1=0
m2=0
v=0
cv=0
c=0
m=0
t=[]
u=[]

print ("\n\nenter X value\n\n")
while True:
    input_x=input('   ')
    if(input_x.isdigit()):
        a.append(float(input_x))
        n=n+1
    else:
        break;

print ("\n\nenter Y value\n\n")
for i in range(0,n):
    input_y=input('   ')
    if(input_y.isdigit()):
        b.append(float(input_y))
    else:
        break;


with tf.name_scope("Scope_FINAL"):
    with tf.name_scope("Mean"):
        n=0
        for i,j in zip(a,b):
            t1=t1+i
            t2=t2+j
            n=n+1
        m1=t1/n
        m2=t2/n
    with tf.name_scope("Scope_C"):
        with tf.name_scope("Scope_Variance"):
            for i,j in zip(a,b):
                w=0
                w=i-m1
                w=w*w
                v=v+w
        with tf.name_scope("Scope_Covariance"):
            for i,j in zip(a,b):
                w=0
                x=0
                w=i-m1
                x=j-m2
                cv=cv+(w*x)
        c=tf.divide(cv, v, name="DIV")
    with tf.name_scope("Scope_M"):
        n=tf.multiply(c, m1, name="MUL")
        m=tf.subtract(m2, n, name="SUB")
    g=tf.multiply(m, 3, name="MUL")
    h=tf.add(g, c, name="ADD")

print("please enter the expected output for numbers 1,2,3,4")
t.append(float(input()))
t.append(float(input()))
t.append(float(input()))
t.append(float(input()))

for i in (1,2,3,4):
    s=m*i+c
    u.append(s)

q=0
for i,j in zip(t,u):
    p=0
    p=j-i
    p=p*p
    q=q+p

q=q/4
r=tf.sqrt(q, name="Sqrt")    
    
    
    
with tf.Session() as sess:
    writer = tf.summary.FileWriter("/tmp/tboard/output", sess.graph)
    print(sess.run(r))
    print(sess.run(h))
    writer.close()
 
        
  
```
