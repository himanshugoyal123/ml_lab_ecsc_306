# For        y=mx+c      Equation

```python
import tensorflow as tf
import matplotlib.pyplot as plt
tf.reset_default_graph()

#take input array
x=[]
y=[]
print ("\npress e to exit \n")
while(1):
    a=input('please provide x value      ')
    b=input('please provide y value      ')
    if(a=='e'):
        print('\nexit succesfull\n')
        break
    
    x.append(float(a))
    y.append(float(b))

    
    
print('\nPlot for x and y input\n')
plt.plot(x, y)
plt.show()



def cvm(x,mx):
        return(x-mx)
    
##variable for variance
varx=0;
vary=0;
#variable for covariance
cv=0;
c=0

    
with tf.name_scope("y_mx_plus_c"):
    with tf.name_scope("Mean_x_y"):
        meanx=tf.reduce_mean(x)
        meany=tf.reduce_mean(y)
    with tf.name_scope("Calculate_C"):
        with tf.name_scope("Variance_x"):
            for i,j in zip(x,y):
                p=0
                p=cvm(i,meanx)
                p=p*p
                varx=varx+p
        ###Calculation for x for covariance
        with tf.name_scope("Covariance_x_y"):
            for i,j in zip(x,y):
                p=0
                q=0
                p=cvm(i,meanx)
                q=cvm(j,meany)
                cv=cv+(p*q)
        c=tf.divide(cv, varx, name="Divide")
    with tf.name_scope("Calculate_M"):
        ## calculate m
        n=tf.multiply(c, meanx, name="MUL")
        m=tf.subtract(meany, n, name="SUB")
    yy=tf.multiply(m, 3, name="MUL_x")
    h=tf.add(yy, c, name="ADD")
    
# To Find the root mean square error
z=[]

for i in x:
    w=m*i+c
    z.append(w)
    u=i


total=0

for i,j in zip(z,y):
    p=0
    p=cvm(j,i)
    p=p*p
    total=total+p

total=total/4
rms=tf.sqrt(total, name="Square_root")


print('\nPlot for Pedicted value vs input value\n')
plt.plot(x, z)
plt.show()


##cacluate y
with tf.Session() as sess:
    writer = tf.summary.FileWriter("/tmp/tboard/output125", sess.graph)
    writer.close()
    print("\nCumulative Frequency\n")
    print(sess.run(cv))
    print("\nM value\n")
    print(sess.run(m))
    print("\nC value\n")
    print(sess.run(c))
    print("\nPredicted value\n")
    print(sess.run(h))
    print("\nRoot Mean Square Value\n")
    print(sess.run(rms))
    
    
```
